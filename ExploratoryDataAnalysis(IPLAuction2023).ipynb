```python
mean=df["base price (in lacs)"].mean()
std=df["base price (in lacs)"].std()
mean,std
```

```python
std1=df["final price (in lacs)"].std()
mean1=df["final price (in lacs)"].mean()
std1,mean1
```

```python
df["zscore"]=(df["base price (in lacs)"]-df["base price (in lacs)"].mean())/df["base price (in lacs)"].std()
df.head()
```

```python
df[df.zscore>3]
```

```python
df_no_outliers=df[(df.zscore<=3) & (df.zscore>=-3)]
df_no_outliers
```

```python
df_no_outliers["zscore_2"]=(df_no_outliers["final price (in lacs)"]-df_no_outliers["final price (in lacs)"].mean())/df_no_outliers["final price (in lacs)"].std()
df_no_outliers
```

```python
df_no_outliers[df_no_outliers.zscore_2>3]
```

```python
df_no_outliers_2=df_no_outliers[(df_no_outliers.zscore_2<=3) & (df_no_outliers.zscore_2>=-3)]
df_no_outliers_2
```
