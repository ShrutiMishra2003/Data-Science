```python
import pandas as pd
from sklearn.datasets import load_iris
import numpy as np
```

```python
iris=load_iris()
dir(iris)
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df['flower'] = iris.target
df['flower'] = df['flower'].apply(lambda x: iris.target_names[x])
df[47:150]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.3)
```

```python
from sklearn.svm import SVC
model=SVC(kernel='rbf',C=30,gamma='auto')
model.fit(X_train,y_train)
```

```python
model.score(X_test, y_test)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
score1=cross_val_score(SVC(kernel='rbf',C=30,gamma='auto'),iris.data,iris.target)
score1
```

```python
np.average(score1)
```

```python
cross_val_score(SVC(kernel='linear',C=20,gamma='auto'),iris.data,iris.target)
```

```python
cross_val_score(SVC(kernel='rbf',C=20,gamma='auto'),iris.data,iris.target)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
clf=GridSearchCV(SVC(gamma='auto'), {
   'C':[1,10,20],
   'kernel':['rbf','linear'],
}, cv=5, return_train_score=False)
clf.fit(iris.data,iris.target)
clf.cv_results_
```

```python
df=pd.DataFrame(clf.cv_results_)
df
```

```python
df[['param_C','param_kernel','mean_test_score']]
```

```python
dir(clf)
```

```python
clf.best_estimator_
```

```python
from sklearn.model_selection import RandomizedSearchCV
rs=RandomizedSearchCV(SVC(gamma='auto'), {
   'C':[1,10,20],
   'kernel':['rbf','linear'],
}, cv=5, n_iter=2, return_train_score=False)
rs.fit(iris.data,iris.target)
```

```python
df1=pd.DataFrame(rs.cv_results_)
df1
```
