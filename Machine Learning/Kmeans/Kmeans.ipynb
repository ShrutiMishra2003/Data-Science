```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.cluster import KMeans
```

```python
df=pd.read_csv("C:\\Users\\Shruti Mishra\\Downloads\\income (1).csv")
df.head()
```

```python
plt.scatter(df.Age,df['Income($)'])
```

```python
km=KMeans(n_clusters=3)
km
```

```python
y_predicted=km.fit_predict(df[['Age','Income($)']])
y_predicted
```

```python
df['cluster']=y_predicted
df.head()
```

```python
df1=df[df.cluster==0]
df1.head()
```

```python
df2=df[df.cluster==1]
df2.head()
```

```python
df3=df[df.cluster==2]
df3.head()
```

```python
plt.scatter(df1.Age,df1['Income($)'])
plt.scatter(df2.Age,df2['Income($)'])
plt.scatter(df3.Age,df3['Income($)'])
plt.xlabel('Age')
plt.ylabel('Income')
plt.legend()
```

```python
scaler=MinMaxScaler()
scaler.fit(df[['Income($)']])
```

```python
df['Income($)']=scaler.transform(df[['Income($)']])
df
```

```python
scaler.fit(df[['Age']])
```

```python
df['Age']=scaler.transform(df[['Age']])
df
```

```python
km=KMeans(n_clusters=3)
km
```

```python
y_predicted1=km.fit_predict(df[['Age','Income($)']])
y_predicted1
```

```python
df['clusters']=y_predicted1
df.drop('cluster',axis='columns',inplace=True)
df.head()
```

```python
df0=df[df.clusters==0]
df0.head()
```

```python
df_1=df[df.clusters==1]
df_1.head()
```

```python
df_2=df[df.clusters==2]
df_2.head()
```

```python
plt.scatter(df0.Age,df0['Income($)'])
plt.scatter(df_1.Age,df_1['Income($)'])
plt.scatter(df_2.Age,df_2['Income($)'])
plt.xlabel('Age')
plt.ylabel('Income')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
```

```python
km.cluster_centers_
```

```python
k_rng= range(1,10)
sse=[]
for k in k_rng:
    km=KMeans(n_clusters=k)
    km.fit(df[['Age','Income($)']])
    sse.append(km.inertia_)
```

```python
sse
```

```python
plt.xlabel('K')
plt.ylabel('SSE')
plt.plot(k_rng,sse)
```
