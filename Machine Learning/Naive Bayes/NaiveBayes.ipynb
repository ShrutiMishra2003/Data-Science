```python
import pandas as pd
```

```python
df=pd.read_csv("C:\\Users\\Shruti Mishra\\Downloads\\titanic (1).csv")
df.head()
```

```python
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns',inplace=True)
df.head()
```

```python
target=df.Survived
target
```

```python
input=df.drop(['Survived'],axis='columns')
input
```

```python
dummies=pd.get_dummies(df.Sex)
dummies=dummies.astype(int)
dummies
```

```python
final=pd.concat([input,dummies],axis='columns')
final
```

```python
final1=final.drop(['Sex'],axis='columns')
final1
```

```python
final1.columns[final1.isna().any()]
```

```python
final1.Age[:10]
```

```python
final1.Age=final1.Age.fillna(final1.Age.mean())
final1
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(final1,target,test_size=0.2)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model=GaussianNB()
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
